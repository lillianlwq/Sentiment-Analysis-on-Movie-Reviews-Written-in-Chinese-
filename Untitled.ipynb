## Import Chinese stopwords from stop_words.txt
convert the stopwords to a non-repeating list format to use later

```python
stop_words_file = open(r"stop_words.txt", 'r', encoding='utf-8')
stopwords = stop_words_file.readlines()
stopwords_list = [i.strip() for i in stopwords]
stopwords_list = list(set(stopwords_list))
print(stopwords_list)
```

## Segment words with Jieba
https://tynbl.github.io/docs/python-xxxy-3/08-05.html

```python
def remove_stopwords(original_text):

    filter_pattern = re.compile('[^\u4E00-\u9FD5]+')
    chinese_only = filter_pattern.sub('', original_text)

    word_list = pseg.cut(chinese_only)
    
    used_flags = ['v', 'a', 'ad']
    meaninful_words = []
    for word, flag in word_list:
        if (word not in stopwords_list) and (flag in used_flags):
            meaninful_words.append(word)
    return ' '.join(meaninful_words)

test_text = extracted.loc[5, 'FILTERED_CONTENT']
print('Original Text：', test_text)
print('\n\nFiltered Text：', remove_stopwords(test_text))
```

```python
class Bi_LSTM_SENTIMENT:
    def __init__(self, training_comment, training_score, testing_comment, testing_score):
        self.model = None
        self.training_comment = training_comment
        self.training_score = training_score
        self.testing_comment = testing_comment
        self.testing_score = testing_score
        
    def train_model(self, epochs):
        self.model = Bi_LSTM_SENTIMENT.build_model()
        self.model.fit(self.training_comment, self.training_score, batch_size=1000, epochs=epochs)
        self.model.save('model/bi_lstm')
        score = self.model.evaluate(self.testing_comment, self.testing_score)
        print(score)
    
    @staticmethod
    def build_model():
        model = Sequential()
        model.add(Embedding(50000, 256, input_length=100))
        model.add(Bidirectional(LSTM(128,implementation=2)))
        model.add(Dropout(0.5))
        model.add(Dense(2, activation='relu'))
        model.compile('RMSprop', 'binary_crossentropy', metrics=['accuracy'])
        return model

    
```
